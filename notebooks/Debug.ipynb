```python
import numpy as np
from typing import Any

ArrayLike = Any

def or_(*args):
    if not args:
        raise ValueError("python_or requires at least one argument")

    return np.array([
        next(v for v in values if v) if any(values) else values[-1]
        for values in zip(*args)
    ])


# ---- Example usage ----
a = np.array([0, 0, 0, 0, 0], dtype=np.int8)
b = np.array([1, 1, -1, 1, 1], dtype=np.int8)

result = or_(a, b)

print("Result:")
print(result)
print("dtype:", result.dtype)
```
